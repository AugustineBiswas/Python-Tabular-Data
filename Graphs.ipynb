```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df
```

used to group by the data according to the days and then find out the mean bill amt and tip for each day.

```python
new_tips_df=tips_df.groupby('day')[['total_bill','tip']].mean()
```

```python
new_tips_df
```

```python
plt.bar(new_tips_df.index,new_tips_df.total_bill);
```

```python
sns.barplot(x='day',y='total_bill',data=tips_df)
```

```python
sns.barplot(data=tips_df,x='day',y='total_bill',hue='sex')
```

```python
flights_df=sns.load_dataset("flights")
flights_df
```

```python
flights_df=flights_df.pivot(index='month',columns='year',values='passengers')
flights_df
```

```python
sns.heatmap(flights_df);
```

```python
plt.title("No. of Passengers(1000s)")
sns.heatmap(flights_df,fmt="d",annot=True,cmap='Blues');
```
